```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\HP\Downloads\ev-data.csv")

car_data = df[df['Vehicle Type'] == 'Bike']
df = df.iloc[:-1]
```

```python
vehicle_counts = df['Vehicle Type'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(vehicle_counts, labels=vehicle_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Vehicle Types')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df.index, df['Battery Capacity (kWh)'])
plt.xlabel('Index')
plt.ylabel('Battery Capacity')
plt.title('Scatter Plot of Battery Capacity')
plt.show()
```

```python
df['Year of Manufacture'] = df['Year of Manufacture'].astype(int)
```

```python
df_with_dummies = pd.get_dummies(df, columns=['Vehicle Type'])
df = pd.concat([df, df_with_dummies], axis=1)
df = df.drop('Vehicle Type', axis=1)
```

```python
df = df.drop('Vehicle Type_Scooter', axis=1)
df = df.drop('Vehicle Type_Bike', axis=1)
```

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
manufacturer_counts = df['Manufacturer'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(manufacturer_counts, labels=manufacturer_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Market Share')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df['Price'].plot(kind='box')
plt.title('Box Plot of Price')
plt.ylabel('Price')
plt.show()
```
